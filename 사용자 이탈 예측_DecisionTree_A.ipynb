## 월 별 로그 데이터 분석
- 사용자의 월 별 접속 횟수가 이탈에 영향을 미치는지 확인

```python
# df_logs의 usedate 컬럼을 datetime 형식으로 변경하세요.
df_logs['usedate'] = pd.to_datetime(df_logs['usedate'])
df_logs
```

```python
# df_logs의 usedate 컬럼을 년월 형태로 표현하여 month 컬럼에 추가하세요. (YYYYMM)
# dt와 strftime 함수 이용
df_logs['month'] = df_logs['usedate'].dt.strftime('%Y%m')
df_logs
```

```python
# df_logs에서 month와 customer_id 별 log_id의 개수를 구하세요.
df_logs_month = df_logs.groupby(['month','customer_id'])[['log_id']].count()
df_logs_month
```

```python
# df_logs_month 인덱스를 초기화 하세요.
df_logs_month = df_logs_month.reset_index()
df_logs_month
```

```python
# df_logs_month 데이터 프레임의 log_id 컬럼의 이름을 freq로 변경하세요
df_logs_month.rename(columns={'log_id':'freq'}, inplace=True)
df_logs_month
```

```python
# df_logs_month 데이터 프레임의 customer_id 별 freq의 최소, 중앙값, 평균, 최대, 합계를 계산하세요.
user_log = df_logs_month.groupby('customer_id')['freq'].agg(['min', 'median', 'mean', 'max', 'sum'])
user_log
```

```python
# user_log 데이터 프레임의 인덱스를 초기화 하세요.
user_log.reset_index(inplace=True)
user_log
```

```python
# user 데이터 프레임과 user_log 데이터 프레임을 userid를 기준으로 병합하세요
user = pd.merge(user, user_log, on='customer_id')
user
```

## 주 별 로그 데이터 분석
- 매주 방문하는 회원 분석

```python
# df_logs의 요일 정보를 wday 컬럼에 추가하세요.
# 0 : 월요일, 6 : 일요일
df_logs['wday'] = df_logs['usedate'].dt.weekday
df_logs
```

```python
# df_logs 데이터 프레임의 customer_id, month, wday 컬럼을 그룹으로 log_id 컬럼의 개수를 구하세요.
df_logs_weekly = df_logs.groupby(['customer_id','month','wday'])['log_id'].agg(['count'])
df_logs_weekly
```

```python
# df_logs_weekly의 인덱스를 초기화 하세요.
df_logs_weekly = df_logs_weekly.reset_index()
df_logs_weekly
```

```python
# 매주 방문하는 사람을 찾기 위해 customer_id 별 count의 최대값을 구하세요.
df_logs_weekly_routine = df_logs_weekly.groupby('customer_id')[['count']].max()
df_logs_weekly_routine
```

```python
# df_logs_weekly_routine 데이터 프레임에 0의 값을 가지는 routine 컬럼을 추가하세요.
df_logs_weekly_routine['routine'] = 0
df_logs_weekly_routine
```

```python
# count 컬럼의 값이 4 이상인 경우 routine 컬럼의 값을 1로 변경하세요
df_logs_weekly_routine.loc[df_logs_weekly_routine['count'] >= 4, 'routine'] = 1
df_logs_weekly_routine
```

```python
# df_logs_weekly_routine 데이터프레임의 인덱스를 초기화하고 customer_id와 routine 컬럼만 선택하세요.
routine = df_logs_weekly_routine.reset_index()[['customer_id', 'routine']]
routine
```

```python
# user 데이터 프레임과 routine 데이터 프레임을 customer_id 컬럼을 기준으로 병합하세요
user = pd.merge(user, routine, on='customer_id')
user
```

```python
# user 데이터 프레임의 routine 컬럼의 고유값 별 개수를 계산하세요
user['routine'].value_counts()
```

```python
# 탈퇴한 회원의 routine 빈도수를 구하세요.
user.loc[user['is_deleted'] == 1, 'routine'].value_counts()
```

```python
# 탈퇴하지 않은 회원의 routine 빈도수를 구하세요.
user.loc[user['is_deleted'] == 0, 'routine'].value_counts()
```

## 이용 기간 구하기

```python
# end_date 컬럼와 start_date 컬럼의 차이를 period 컬럼에 추가하세요.
user['period'] = user['end_date'] - user['start_date']
user
```

```python
# period 의 일 정보를 period_day 컬럼에 추가하세요.
user['period_day'] = user['period'].dt.days
user
```

## ML (Classification)

```python
# user 데이터 프레임을 조회하세요.
user
```

```python
# user 데이터 프레임에서 다음의 결과와 같이 컬럼을 선택 후 df 변수에 저장하세요.
df = user[['gender', 'is_deleted', 'class_name', 'price', 'campaign_name', 'min', 'median', 'mean', 'max', 'sum', 'routine', 'period_day']]
df
```

```python
# df에서 is_deleted 컬럼을 Y로, 나머지를 X로 설정하세요
X = df.drop('is_deleted', axis = 1)
Y = df['is_deleted']
```

```python
# X를 출력하세요.
X
```

```python
# Y를 출력하세요.
Y
```

```python
# X를 One Hot 변환하세요
X = pd.get_dummies(X)
X
```

```python
# X와 Y를 Train 과 Test로 분할하세요.
# stratify=Y, random_state=0
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, stratify=Y, random_state=0)
```

```python
# KNN 알고리즘을 이용하여 분류하세요.
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3).fit(x_train, y_train)
model.score(x_train, y_train), model.score(x_test, y_test)
```

```python
# LogisticRegression 알고리즘을 이용하여 분류하세요.
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=10000).fit(x_train, y_train)
model.score(x_train, y_train), model.score(x_test, y_test)
```

```python
# DecisionTree 알고리즘을 이용하여 분류하세요.
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=6).fit(x_train, y_train)
model.score(x_train, y_train), model.score(x_test, y_test)
```

```python
# 학습된 DecisionTree 모델의 feature_importances_를 Series로 변환하세요.
fi = pd.Series(model.feature_importances_, index=X.columns)
fi
```

```python
# 변환된 결과를 시각화 하세요.
fi.sort_values(ascending=True).plot(kind='barh')
plt.show()
```

```python
# 트리를 시각화하세요.
# max_depth = 2,
# feature_names = X.columns
# fontsize = 16
# filled = True
# rounded = True

from sklearn.tree import plot_tree
plt.figure(figsize = (20,20))
plot_tree(model,
          max_depth = 2,
          feature_names = X.columns,
          fontsize = 16,
          filled = True,
          rounded = True)
plt.show()
```
